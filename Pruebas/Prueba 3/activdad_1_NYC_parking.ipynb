````python
# Cargamos el paquete de pandas:
import pandas as pd
````

````python
# Cargamos el fichero de datos, con la peculiaridad que se emplea como separador ";".
df = pd.read_csv("Datos/sample_NYC_parking_issues.csv", sep = ";")

# Visualizamos los datos para comprobar que mantiene una estructura coherente:
df.head()
````


#### A, ¿Qué columnas no aportan ningún valor a las muestras?


````python
# Primeramente observamos de cuantas filas se compone el dataframe:
df.shape

# Ahora filtramos aquellas columnas con todos los valores NaN y vemos que las 3 últimas columnas en concreto se encuentran vacías:
df.isna().all()

# Por lo que, al tener la misma longitud que las columnas, lo aplicamos y obtenemos el vector con las columnas que no aportan ningún valor:
df.columns[df.isna().all()]
````

####  B. Selecciona las muestras "4DSD" en `Vehicle Body Type`  y guardalas en un nuevo fichero


````python
# Filtramos la columna por el string indicado y lo indexamos de nuevo en el dataframe, guardándolo como un nuevo df:
df2 = df[df["Vehicle Body Type"] == "4DSD"]

# Exportamos como un nuevo fichero csv el dataframe filtrado:
df2.to_csv("Datos/Vehicle Body Type.csv")
````

#### C. ¿Cuántos colores de vehículos hay?


````python
# A continuación aplicamos el método value_counts() para agrupar por valores únicos y medimos la longitud de la serie, dando el total de colores:
count_colores = len(df["Vehicle Color"].value_counts())

print("El total de colores registrados es de " + str(count_colores))
````

#### D. ¿Cuál es el rango de años en la antigüedad del vehículo? 


````python
# Suponemos que la antigüedad viene dada en la columna del año del vehículo, por lo que buscamos el año máximo y el mínimo:
minimo = df["Vehicle Year"].min()
maximo = df["Vehicle Year"].max()

print("El rango de años dado es de " + str(int(minimo)) + "-" + str(int(maximo)))
````

#### E. ¿Cuál es el tipo de vehículo más viejo?


````python
# Filtramos los datos por el valor mínimo encontrado anteriormente, escogemos la columna, aplicamos value_counts:
df[df["Vehicle Year"] == minimo]["Vehicle Body Type"].value_counts()

# Vemos que existen varios tipos de vehículos que cumplen la condición de tener una antigüedad igual a la mínima, por lo que el criterio para
# decidir que vehículo es el más viejo será escoger el que mayor número de ellos tiene registrados:
mas_antiguo = df[df["Vehicle Year"] == minimo]["Vehicle Body Type"].value_counts().index[0]

print("El tipo de vehículo con un mayor número de registros en el año mínimo es " + str(mas_antiguo))
````

####  F. ¿Cuál es la distancia mínima del vehículo a la acera "Feet From Curb"?


````python
# Del enunciado entiendo que simplemente hay que encontrar el valor mínimo en la columna indicada:
df["Feet From Curb"].min()
````

####  G. Selecciona aquellas muestras que estén por debajo de la distancia `Feet From Curb` media


````python
# Primero calculamos la distancia media de la columna indicada:
media = df["Feet From Curb"].mean()

# Ahora filtramos aquellas filas cuya distancia en la columna indicada es estrictamente menor a la media:
df[df["Feet From Curb"] < media]
````

####  H. Elimina las dos últimas columnas del dataframe creando uno nuevo.


````python
# Podemos eliminar las dos últimas columnas a partir de un slicing:
df3 = df.iloc[:,:-2]
````

#### I. Selecciona la 4º, 6º y 8º columna de las filas comprendidas entre la 1 y 100, y la 200 y 500.


````python
# Al empezar los índices en 0, seleccionamos las columnas con el índice i-1, ya que la cuarta columna en realidad sería aquella cuyo índice
# es 3. Mientras que, las filas sería lo mismo, seleccionando el índice i-1:
c1 = df.iloc[0:100,[3,5,7]]
c2 = df.iloc[199:500,[3,5,7]]

# Ahora concatenamos ambos rangos de filas:
df4 = pd.concat([c1,c2])
````

#### J. ¿Hay vehículos que han recibido más de una multa ?


````python
repetidores = len(df["Plate ID"].value_counts() > 1)

print("Sí, en concreto del total de 10000 registros, " + str(repetidores) + " matrículas se repiten como mínimo más de una vez")
````

####  K. Los nombres de los propietarios (nombres ficticios) están en otro fichero (`plates_properties.csv`), combina ambos ficheros con coherencia en relación a la informació y guarda el resultado en otro documento csv.

````python
df_merge = pd.read_csv("Datos/plates_properties.csv", index_col=0)

# Creamos un merge left, donde queremos mantener los datos del dataframe trabajado, sumado a los nombres de correspondientes a las matrículas
# de coche que se tienen.
df5 = df.merge(df_merge, left_on="Plate ID", right_on="plate_id", how="left")

# Guardamos en un archivo csv:
df5.to_csv("Datos/Ejercicio K.csv")
````
